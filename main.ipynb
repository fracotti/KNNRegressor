## Створимо KNNRegressor власноруч

- Евклідова метрика<br>
&emsp;Відстані між двома точками $p = (p_1, p_2, ..., p_n)$ та $q = (q_1, q_2, ..., q_n)$<br>
&emsp;у n-вимірному просторі виглядає наступним чином:
<br>
<br>

$$d(p, q) = \sqrt{\sum_{i=1}^{n}(p_i - q_i)^2}$$


```python
def euclidean_distance(X, Y):
    XX = np.sum(X**2, axis=1, keepdims=True)
    YY = np.sum(Y**2, axis=1)
    XY = np.dot(X, Y.T)
    return np.sqrt(XX - 2*XY + YY)
```

```python
class KNNRegressorDummy:
    def __init__(self, X, y, n_neighbors=10):
        self.X = X
        self.y = y
        self.k = n_neighbors

    def predict(self, X_test):
        y_pred = []
        for x in X_test:
            distances = euclidean_distance(np.array([x]), self.X)[0]
            y_pred.append(self.y[np.argsort(distances)[:self.k]].mean())
        return y_pred
```

```python
r2_all_scaled_custom = []
for k in range(1, 21):
    custom_knn_regressor = KNNRegressorDummy(X_train, y_train, n_neighbors=k)

    test_pred = custom_knn_regressor.predict(X_test)

    test_r2_score = r2_score(test_pred, y_test)
    r2_all_scaled_custom.append(test_r2_score)

    print('-'*20, 'K =', k, '-'*20)
    print(f'KNN Regression R^2 is {test_r2_score}')
```

```python
r2_df_custom_scaled = pd.DataFrame(
    {
        'KNN k': np.arange(1, 21),
        'Accuracy' : r2_all_scaled_custom
    }
)
r2_df_custom_scaled
```
